```python
 import nltk
from nltk.chat.util import Chat, reflections

pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, How are you today ?",]
    ],
    [
        r"what is your name ?",
        ["My name is Chatty and I'm a chatbot !"]
    ],
    [
        r"how are you ?",
        ["I'm fine, thank you! How about you ?",]
    ],
    [
        r"exit",
        ["Bye for now, see you soon !",]
    ]
]

chat = Chat(pairs, reflections)
chat.converse()

```
